```python
import pandas as pd

mei_df = pd.read_csv("../data/processed/mei.csv")
rain_df = pd.read_csv("../data/processed/colombo_rainfall.csv")
temp_df = pd.read_csv("../data/processed/colombo_temperature.csv")

mei_df["Date"] = pd.to_datetime(mei_df["Date"])
rain_df["Date"] = pd.to_datetime(rain_df["Date"])
temp_df["Date"] = pd.to_datetime(temp_df["Date"])
```

```python
climate_df = (
    rain_df
    .merge(temp_df, on="Date")
    .merge(mei_df, on="Date")
)

print(climate_df.shape)

climate_df.head()
```

```python
climate_df.to_csv(
    "../data/processed/climate_dataset.csv",
    index=False
)

print("Climate dataset saved!")
```
