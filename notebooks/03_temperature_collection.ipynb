```python
import pandas as pd
import requests

LAT = 6.9271
LON = 79.8612

url = (
    "https://power.larc.nasa.gov/api/temporal/monthly/point"
    "?parameters=T2M"
    "&community=AG"
    "&longitude=79.8612"
    "&latitude=6.9271"
    "&start=2000"
    "&end=2025"
    "&format=JSON"
)

response = requests.get(url)

print(response.status_code)
```

```python
temp_data = (
    response.json()
    ["properties"]
    ["parameter"]
    ["T2M"]
)

list(temp_data.items())[:5]
```

```python
temp_df = pd.DataFrame(
    temp_data.items(),
    columns=["Date", "Temperature"]
)

temp_df.head()
```

```python
temp_df = temp_df[
    ~temp_df["Date"].str.endswith("13")
].copy()

print(temp_df.shape)
```

```python
temp_df["Date"] = pd.to_datetime(
    temp_df["Date"],
    format="%Y%m"
)

temp_df.info()
```

```python
temp_df.to_csv(
    "../data/processed/colombo_temperature.csv",
    index=False
)

print("Temperature dataset saved!")
```
